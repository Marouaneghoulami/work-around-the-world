```python
from pyspark.sql import SparkSession
import configparser

config = configparser.ConfigParser()
config.read_file(open('dc.cfg'))

landingJobsApiKey = config.get('LANDINGJOBS', 'API_KEY')

spark = SparkSession.builder.appName("Truck Data Wrangler").getOrCreate()
spark
```

```python
import requests as r
import pandas as pd

# https://github.com/LandingJobs/LandingJobs-api

lj_jobs_offset = 0
lj_jobs_offset_interval = 50
lj_jobs_list = []

while lj_jobs_offset < 500:
    params = {'offset': lj_jobs_offset}
    response = r.get('https://landing.jobs/api/v1/jobs', params)
    lj_jobs_list.extend(response.json())
    lj_jobs_offset = lj_jobs_offset + lj_jobs_offset_interval


jobs_list_df = pd.DataFrame(lj_jobs_list)
jobs_list_df
```

```python
response = r.get('https://landing.jobs/api/v1/companies')
lj_companies_list = response.json()
len(lj_companies_list)
```
