### Landing Jobs (API)

Source: https://github.com/LandingJobs/LandingJobs-api

```python
import requests as r
import pandas as pd

lj_jobs_offset = 0
lj_jobs_offset_interval = 50
lj_jobs_list = []

while lj_jobs_offset < 500:
    params = {'offset': lj_jobs_offset}
    response = r.get('https://landing.jobs/api/v1/jobs', params)
    lj_jobs_list.extend(response.json())
    lj_jobs_offset = lj_jobs_offset + lj_jobs_offset_interval


jobs_list_df = pd.DataFrame(lj_jobs_list)
jobs_list_df
```

```python
response = r.get('https://landing.jobs/api/v1/companies')
lj_companies_list = response.json()
len(lj_companies_list)
```

### Jobtechdev.se historical jobs

Source: https://jobtechdev.se/api/jobs/historical/

```python
import pandas as pd

def translate_jobtechdev_columns(df):
    return (
        df.withColumnRenamed('AG_NAMN', 'agency_name')
                   .withColumnRenamed('ANTAL_AKT_PLATSER', 'number_of_acting_sites')
                   .withColumnRenamed('ARBETSDRIFT', 'work_period')
                   .withColumnRenamed('ARBETSTID', 'working_hours')
                   .withColumnRenamed('BESKR_ARBETSDRIFT', 'working_hours_description')
                   .withColumnRenamed('FORSTA_PUBLICERINGSDATUM', 'published_at')
                   .withColumnRenamed('HEMSIDA', 'website')
                   .withColumnRenamed('KOMMUN_KOD', 'community_code')
                   .withColumnRenamed('ORGNR', 'company_number')
                   .withColumnRenamed('PLATSBESKRIVNING', 'job_description')
                   .withColumnRenamed('PLATSNUMMER', 'job_number')
                   .withColumnRenamed('PLATSRUBRIK', 'job_title')
                   .withColumnRenamed('POSTNR', 'zipcode')
                   .withColumnRenamed('POSTORT', 'city')
                   .withColumnRenamed('SISTA_ANSOK_PUBLDATUM', 'expires_at')
                   .withColumnRenamed('TILLTRADE', 'urgency_level_description')
                   .withColumnRenamed('VARAKTIGHET', 'duration')
    )

jtd_hist = 'datasets/jobtechdev-historical/2008.json'
jtd_hist_df = spark.read.json(jtd_hist)
jtd_hist_df = translate_jobtechdev_columns(jtd_hist_df)
jtd_hist_df.limit(100).toPandas()
```

### GitHub Jobs API

Source: https://jobs.github.com/api

```python
import requests as r
import pandas as pd

github_jobs_api_url = 'https://jobs.github.com/positions.json'

params = {'page': 1}

response = r.get(github_jobs_api_url, params)

results = response.json()
results_buffer = []
results_buffer.extend(results)

while len(results) > 0:
    params['page'] = params['page'] + 1
    response = r.get(github_jobs_api_url, params)
    results = response.json()
    if len(results) > 0:
        results_buffer.extend(results)
    else:
        break

jobs_df = pd.DataFrame(results_buffer)
jobs_df
```
